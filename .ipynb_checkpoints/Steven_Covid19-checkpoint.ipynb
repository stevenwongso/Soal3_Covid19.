```python
from bs4 import BeautifulSoup
import requests
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import folium
```

```python
url = 'https://en.wikipedia.org/wiki/2020_coronavirus_pandemic_in_Indonesia'
html = requests.get(url)
soup = BeautifulSoup(html.content, 'html.parser')
findtable =str(soup.find('table', {"class":"wikitable float sortable"}))
soup = BeautifulSoup(findtable,'html.parser' )
findvalue = soup.find_all('td')

indexconfirmed=[0]
indexrecovered=[1]
indexdeath=[2]
indexactive=[3]

confirmed=0
recovered=1
death=2
active=3

dataconfirmed=[]
datarecovered=[]
datadeath=[]
dataactive=[]

for i in findvalue:
    confirmed += 5
    recovered += 5
    death += 5
    active += 5
    indexconfirmed.append(confirmed)
    indexrecovered.append(recovered)
    indexdeath.append(death)
    indexactive.append(active)

for i in range(len(findvalue)) :
    if i in indexconfirmed :
        dataconfirmed.append(findvalue[i].text.replace('\n',''))
    elif i in indexrecovered :
        datarecovered.append(findvalue[i].text.replace('\n',''))
    elif i in indexdeath :
        datadeath.append(findvalue[i].text.replace('\n',''))
    elif i in indexactive :
        dataactive.append(findvalue[i].text.replace('\n',''))
        
dataconfirmed.pop()

findprovince = soup.find_all('th')

dataprovince = []

for i in range(11, len(findprovince)-7):
    dataprovince.append(findprovince[i].text.replace('\n','').replace('[b]',''))

data=[]
for i in range(len(dataprovince)):
    data.append({
        "province": dataprovince[i],
        "confirmed" : dataconfirmed[i],
        "recovered" : datarecovered[i],
        "death" : datadeath[i],
        "active" : dataactive[i]
    })
```

```python
dfcovid=pd.DataFrame(data)
dfcovid
```

```python
url2='https://raw.githubusercontent.com/LintangWisesa/Indonesia-Covid19-Maps/master/data/gps_indonesia.json'
data2 = requests.get(url2)
hasil = data2.json()
dfgeo = pd.DataFrame(hasil)
dfgeo
```

```python
peta = folium.Map(
    location = [-2.2756707, 116.4386513],
    tiles = 'OpenStreetMap',
    zoom_start = 5
)
for i in range(len(dfcovid)):
    folium.Marker(
        [dfgeo.iloc[i]['latitude'],dfgeo.iloc[i]['longitude']],
        popup = f"<b><h3> {dfcovid.iloc[i]['province']} </h3> </b> <br> Confirmed : {dfcovid.iloc[i]['confirmed']} <br> Recovered : {dfcovid.iloc[i]['recovered']}  <br>  Death : {dfcovid.iloc[i]['death']}  <br>  Active: {dfcovid.iloc[i]['active']}",
        icon = folium.Icon(),
        tooltip = 'Klik untuk detail kasus'
    ).add_to(peta)
    
peta.save('Steven_Covid19.html')
peta
```
